#Split gmaps_link



*   lat

```python
# make new column 'lat' from gmaps_link
trails_ca['lat'] = trails_ca['gmaps_link'].str.split('=').str[1].str.split(',').str[0] 
# make new column 'lon' from gmaps_link
trails_ca['lon'] = trails_ca['gmaps_link'].str.split('=').str[1].str.split(',').str[1] 
```

```python
# trim trail_name
trails_ca['trail_name'] = trails_ca['trail_name'].str.strip()
# trim location
trails_ca['location'] = trails_ca['location'].str.strip()
# trim high
trails_ca['high'] = trails_ca['high'].str.strip()
# trim low
trails_ca['low'] = trails_ca['low'].str.strip()
```

```python
trails_ca.head()
```

```python
import re
# # as per recommendation from @freylis, compile once only
def remove_tags(string):
    result = re.sub('<.*?>','',string)
    return result
```

```python
# # trim description
# trails_bc['description'] = cleanhtml(str(trails_bc['description']))
trails_ca['description'] = trails_ca['description'].apply(lambda cw : remove_tags(str(cw)))
trails_ca['description'] = trails_ca['description'].str.strip()
```

```python
trails_ca.head()
```

```python
trails_ca['rank'].fillna(0).astype(int)
trails_ca['rating'].fillna(0).astype(float)
trails_ca['num_reviews'].fillna(0).astype(int)
trails_ca['length'].fillna(0).astype(float)
trails_ca['high'] = trails_ca['high'].str.replace(',', '')
trails_ca['high'].fillna(0).astype(int)
trails_ca['low'] = trails_ca['low'].str.replace(',', '')
trails_ca['low'].fillna(0).astype(int)
trails_ca['lat'].fillna(0).astype(float)
trails_ca['lon'].fillna(0).astype(float)
```

# Add suggested_duration

```python
# use a scalar to create time for column
# add duration based on length

trails_ca_filtered['suggested_duration'] = ''

for index in trails_ca_filtered.index: 
  trails_ca_filtered['suggested_duration'][index] = str((trails_ca_filtered['length'][index] / 3).round(1)) + ' hours'
```

# Create numeric value for difficulty rating

```python
trails_ca_filtered['numeric_difficulty'] = ''

for index in trails_ca_filtered.index:
  
  if (trails_ca_filtered['difficulty'][index] == 'Very Difficult'):
    trails_ca_filtered['numeric_difficulty'][index] = 6
  elif (trails_ca_filtered['difficulty'][index] == 'Difficult'):
    trails_ca_filtered['numeric_difficulty'][index] = 5
  elif (trails_ca_filtered['difficulty'][index] == 'Intermediate/Difficult'):
    trails_ca_filtered['numeric_difficulty'][index] = 4
  elif (trails_ca_filtered['difficulty'][index] == 'Intermediate'):
    trails_ca_filtered['numeric_difficulty'][index] = 3
  elif (trails_ca_filtered['difficulty'][index] == 'Easy/Intermediate'):
    trails_ca_filtered['numeric_difficulty'][index] = 2
  else: 
    trails_ca_filtered['numeric_difficulty'][index] = 1
```

```python
trails_ca_filtered.head()
```

# Create duration_low and duration_high

```python
trails_ca_filtered['duration_low'] = ''
trails_ca_filtered['duration_high'] = ''


for index in trails_ca_filtered.index:
  
  if (trails_ca_filtered['difficulty'][index] == 'Very Difficult'):
    trails_ca_filtered['duration_low'][index] = (trails_ca_filtered['length'][index] / 3).round(2)
    trails_ca_filtered['duration_high'][index] = ((trails_ca_filtered['length'][index] / 3) * 2).round(2)
  if (trails_ca_filtered['difficulty'][index] == 'Difficult'):
    trails_ca_filtered['duration_low'][index] = (trails_ca_filtered['length'][index] / 3).round(2)
    trails_ca_filtered['duration_high'][index] = ((trails_ca_filtered['length'][index] / 3) * 1.5).round(2)
  if (trails_ca_filtered['difficulty'][index] == 'Intermediate/Difficult'):
    trails_ca_filtered['duration_low'][index] = (trails_ca_filtered['length'][index] / 3).round(2)
    trails_ca_filtered['duration_high'][index] = ((trails_ca_filtered['length'][index] / 3) * 1.25).round(2)
  if (trails_ca_filtered['difficulty'][index] == 'Intermediate'):
    trails_ca_filtered['duration_low'][index] = (trails_ca_filtered['length'][index] / 3).round(2)
    trails_ca_filtered['duration_high'][index] = (trails_ca_filtered['length'][index] / 3).round(2)
  if (trails_ca_filtered['difficulty'][index] == 'Easy/Intermediate'):
    trails_ca_filtered['duration_low'][index] = ((trails_ca_filtered['length'][index] / 3) * .75).round(2)
    trails_ca_filtered['duration_high'][index] = (trails_ca_filtered['length'][index] / 3).round(2) 
  if (trails_ca_filtered['difficulty'][index] == 'Easy'):
    trails_ca_filtered['duration_low'][index] = ((trails_ca_filtered['length'][index] / 3) * .5).round(2)
    trails_ca_filtered['duration_high'][index] = (trails_ca_filtered['length'][index] / 3).round(2) 
```
