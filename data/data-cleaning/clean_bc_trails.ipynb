#Split gmaps_link



*   lat
*   lon





```python
# make new column 'lat' from gmaps_link
trails_bc['lat'] = trails_bc['gmaps_link'].str.split('=').str[1].str.split(',').str[0] 
# make new column 'lon' from gmaps_link
trails_bc['lon'] = trails_bc['gmaps_link'].str.split('=').str[1].str.split(',').str[1] 
```

```python
trails_bc.head()
```

```python
# trim trail_name
trails_bc['trail_name'] = trails_bc['trail_name'].str.strip()
# trim location
trails_bc['location'] = trails_bc['location'].str.strip()
# trim high
trails_bc['high'] = trails_bc['high'].str.strip()
# trim low
trails_bc['low'] = trails_bc['low'].str.strip()
```

```python
trails_bc.head()
```

```python
import re
# # as per recommendation from @freylis, compile once only
def remove_tags(string):
    result = re.sub('<.*?>','',string)
    return result
```

```python
# # trim description
# trails_bc['description'] = cleanhtml(str(trails_bc['description']))
trails_bc['description'] = trails_bc['description'].apply(lambda cw : remove_tags(str(cw)))
trails_bc['description'] = trails_bc['description'].str.strip()
```

```python
trails_bc.head()
```

```python
trails_bc['rank'].fillna(0).astype(int)
trails_bc['rating'].fillna(0).astype(float)
trails_bc['num_reviews'].fillna(0).astype(int)
trails_bc['length'].fillna(0).astype(float)
trails_bc['high'] = trails_bc['high'].str.replace(',', '')
trails_bc['high'].fillna(0).astype(int)
trails_bc['low'] = trails_bc['low'].str.replace(',', '')
trails_bc['low'].fillna(0).astype(int)
trails_bc['lat'].fillna(0).astype(float)
trails_bc['lon'].fillna(0).astype(float)
```

#Add suggested_duration


```python
# use a scalar to create time for column
# add duration based on length

trails_bc_filtered['suggested_duration'] = ''

for index in trails_bc_filtered.index: 
  trails_bc_filtered['suggested_duration'][index] = str((trails_bc['length'][index] / 3).round(1)) + ' hours'
  
```

#Create numeric difficulty column

```python
trails_bc_filtered['numeric_difficulty'] = ''

for index in trails_bc_filtered.index:
  
  if (trails_bc_filtered['difficulty'][index] == 'Very Difficult'):
    trails_bc_filtered['numeric_difficulty'][index] = 6
  elif (trails_bc_filtered['difficulty'][index] == 'Difficult'):
    trails_bc_filtered['numeric_difficulty'][index] = 5
  elif (trails_bc_filtered['difficulty'][index] == 'Intermediate/Difficult'):
    trails_bc_filtered['numeric_difficulty'][index] = 4
  elif (trails_bc_filtered['difficulty'][index] == 'Intermediate'):
    trails_bc_filtered['numeric_difficulty'][index] = 3
  elif (trails_bc_filtered['difficulty'][index] == 'Easy/Intermediate'):
    trails_bc_filtered['numeric_difficulty'][index] = 2
  else: 
    trails_bc_filtered['numeric_difficulty'][index] = 1
```

```python
trails_bc_filtered.head(30)
```

# create duration_low and duration_high

```python
trails_bc_filtered['duration_low'] = ''
trails_bc_filtered['duration_high'] = ''


for index in trails_bc_filtered.index:
  
  if (trails_bc_filtered['difficulty'][index] == 'Very Difficult'):
    trails_bc_filtered['duration_low'][index] = (trails_bc_filtered['length'][index] / 3).round(2)
    trails_bc_filtered['duration_high'][index] = ((trails_bc_filtered['length'][index] / 3) * 2).round(2)
  if (trails_bc_filtered['difficulty'][index] == 'Difficult'):
    trails_bc_filtered['duration_low'][index] = (trails_bc_filtered['length'][index] / 3).round(2)
    trails_bc_filtered['duration_high'][index] = ((trails_bc_filtered['length'][index] / 3) * 1.5).round(2)
  if (trails_bc_filtered['difficulty'][index] == 'Intermediate/Difficult'):
    trails_bc_filtered['duration_low'][index] = (trails_bc_filtered['length'][index] / 3).round(2)
    trails_bc_filtered['duration_high'][index] = ((trails_bc_filtered['length'][index] / 3) * 1.25).round(2)
  if (trails_bc_filtered['difficulty'][index] == 'Intermediate'):
    trails_bc_filtered['duration_low'][index] = (trails_bc_filtered['length'][index] / 3).round(2)
    trails_bc_filtered['duration_high'][index] = (trails_bc_filtered['length'][index] / 3).round(2)
  if (trails_bc_filtered['difficulty'][index] == 'Easy/Intermediate'):
    trails_bc_filtered['duration_low'][index] = ((trails_bc_filtered['length'][index] / 3) * .75).round(2)
    trails_bc_filtered['duration_high'][index] = (trails_bc_filtered['length'][index] / 3).round(2) 
  if (trails_bc_filtered['difficulty'][index] == 'Easy'):
    trails_bc_filtered['duration_low'][index] = ((trails_bc_filtered['length'][index] / 3) * .5).round(2)
    trails_bc_filtered['duration_high'][index] = (trails_bc_filtered['length'][index] / 3).round(2) 
```
