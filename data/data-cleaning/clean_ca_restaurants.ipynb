

#Parse city and category, and cast columns





```python
# Create 'category' column
restaurants_ca['category'] = restaurants_ca['rank_in_city'].str.split(' ').str[3]

# Rename 'categories' to 'tags and comma-separate
restaurants_ca.rename(columns={'categories':'tags'}, inplace=True)

# Create 'city' and 'state' columns
restaurants_ca['city'] = restaurants_ca['rank_in_city'].str.split(' in ').str[1]
restaurants_ca['state'] = 'CA'

# # Create 'rank_city' and 'num_attractions_city' columns and cast to int
# restaurants_ca['rank_city'] = restaurants_ca['rank_in_city'].str.split(' ').str[0].str.strip('#').str.replace(',', '')
# restaurants_ca['rank_city'].fillna(0).astype(int) # NaN is converted to 0
# restaurants_ca['num_attractions_city'] = restaurants_ca['rank_in_city'].str.split(' ').str[2].str.replace(',', '')
# restaurants_ca['num_attractions_city'].fillna(0).astype(int) # NaN is converted to 0
restaurants_ca.drop('rank_in_city', axis=1, inplace=True)

# Cast column 'num_reviews' to type int
restaurants_ca['num_reviews'] = restaurants_ca['num_reviews'].str.replace(',', '')
restaurants_ca['num_reviews'].fillna(0).astype(int)

# Cast column 'rating' to type int
restaurants_ca['rating'] = restaurants_ca['rating'].str.split(' ').str[0]
restaurants_ca['rating'].astype(float)
```

# Re-categorize



*   Restaurant
*   Cafe
*   Bar
*   Quick Bites
*   Bakery
*   Specialty Food Market





```python
restaurants_ca['temp'] = restaurants_ca['name'] + ' ' + restaurants_ca['category'] + ' ' + restaurants_ca['tags']

restaurants_ca['category'] = np.nan

cafe = ['cafe', 'coffee']
bar = ['bar', 'pub', 'wine']
bakery = ['dessert', 'pie', 'cake', 'bakery']
quick_bites = ['quick bites', 'fast food', 'speciality']
brewery = ['brewery']

for index in restaurants_ca.index:
  
  if any(keyword in str(restaurants_ca['temp'][index]).lower() for keyword in cafe):
    restaurants_ca['category'][index] = 'Cafe'
  elif any(keyword in str(restaurants_ca['temp'][index]).lower() for keyword in brewery):
    restaurants_ca['category'][index] = 'Brewery'
  elif any(keyword in str(restaurants_ca['temp'][index]).lower() for keyword in bar):
    restaurants_ca['category'][index] = 'Bar'
  elif any(keyword in str(restaurants_ca['temp'][index]).lower() for keyword in bakery):
    restaurants_ca['category'][index] = 'Bakery'
  elif any(keyword in str(restaurants_ca['temp'][index]).lower() for keyword in quick_bites):
    restaurants_ca['category'][index] = 'Quick bites'
  else:
    restaurants_ca['category'][index] = 'Restaurant'

restaurants_ca.drop('temp', axis=1, inplace=True)
```

# Split cost to cost_low and cost_high; delete cost

```python
restaurants_ca['cost_low'] = restaurants_ca['cost'].str.split(' ').str[0].str.len()
restaurants_ca['cost_low'].fillna(0).astype(int) # NaN is converted to 0

# this sets cost_high to be the second set of dollar signs
# if there is not a second set of dollar signs, cost_high gets set to what cost_low is
restaurants_ca['cost_high'] = np.where(restaurants_ca['cost'].str.split('- ').str[1].str.len() > 1, restaurants_ca['cost'].str.split('- ').str[1].str.len(), restaurants_ca['cost'].str.split('- ').str[0].str.len())
restaurants_ca['cost_high'].fillna(0).astype(int) # NaN is converted to 0

restaurants_ca.drop('cost', axis=1, inplace=True)
```

#Add suggested_duration

```python
# add duration based on cost_high

restaurants_ca['suggested_duration'] = ''

for index in restaurants_ca.index:
  
  if (restaurants_ca['cost_high'][index] == 4.0):
    restaurants_ca['suggested_duration'][index] = '2-3 hours'
  elif (restaurants_ca['cost_high'][index] == 3.0):
    restaurants_ca['suggested_duration'][index] = '1-2 hours'
  elif (restaurants_ca['cost_high'][index] == 2.0):
    restaurants_ca['suggested_duration'][index] = '1-2 hours'
  elif (restaurants_ca['cost_high'][index] == 1.0):
    restaurants_ca['suggested_duration'][index] = '< 1 hour'
  else: 
    restaurants_ca['suggested_duration'][index] = '1-2 hours'
```

```python
# cases: '1-2 hours', '2-3 hours', '< 1 hour'
restaurants_ca['duration_low'] = ''
restaurants_ca['duration_high'] = ''


for index in restaurants_ca.index:
  
  if (restaurants_ca['suggested_duration'][index] == '< 1 hour'):
    restaurants_ca['duration_low'][index] = .5
    restaurants_ca['duration_high'][index] = 1.0
  if (restaurants_ca['suggested_duration'][index] == '1-2 hours'):
    restaurants_ca['duration_low'][index] = 1.0
    restaurants_ca['duration_high'][index] = 2.0
  if (restaurants_ca['suggested_duration'][index] == '2-3 hours'):
    restaurants_ca['duration_low'][index] = 2.0
    restaurants_ca['duration_high'][index] = 3.0
```
